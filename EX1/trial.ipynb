```python
import json
import matplotlib.pyplot as plt

# 解析日志（18_adam）
train_data = []
val_data = []
with open('/Users/niezhiqi/Desktop/课程/深度学习/mmpretrain/work_dirs/res18_adam/20250510_114733/vis_data/20250510_114733.json', 'r') as f:
    for line in f:
        entry = json.loads(line.strip())
        if 'accuracy/top1' in entry:  # 验证集数据
            val_data.append(entry)
        elif 'loss' in entry and 'accuracy/top1' not in entry:  # 训练集数据
            train_data.append(entry)

# 提取关键指标
train_steps = [x['step'] for x in train_data]
train_loss = [x['loss'] for x in train_data]
val_steps = [x['step'] for x in val_data]
val_acc = [x['accuracy/top1'] for x in val_data]
#绘制acc图
plt.figure(figsize=(10, 5))
plt.plot(val_steps, val_acc, label='Validation Accuracy')
plt.xlabel('Steps')
plt.ylabel('Accuracy')
plt.title('Validation Accuracy over Steps')
plt.legend()
plt.grid()
plt.savefig('/Users/niezhiqi/Desktop/figure/accuracy_plot18_adam.png')
plt.show()



```

```python
#resnet18+adam
import json
import matplotlib.pyplot as plt

# 1. 读取 JSON 文件
def load_json_data(json_path):
    with open(json_path, 'r') as f:
        data = json.load(f)
    return data

# 2. 绘制准确率曲线
def plot_accuracy_curves(json_path, save_path=None):
    # 加载数据
    data = load_json_data(json_path)
    
    # 提取训练和验证准确率
    val_acc = data['val_acc_history']
    epochs = range(1, len(val_acc) + 1)
    
    # 绘制图像
    plt.figure(figsize=(10, 5))
    plt.plot(epochs, val_acc, 'r-', label='Validation Accuracy')
    
    # 设置标题和标签
    plt.title(' Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True)
    
    # 保存图像（可选）
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
    
    plt.show()

# 3. 调用函数绘制图像
json_path = '/Users/niezhiqi/Desktop/课程/深度学习/mmpretrain/work_dirs/res18_adam/20250510_114733/vis_data/20250510_114733.json'  # 替换为你的 JSON 文件路径
plot_accuracy_curves(json_path, save_path='/Users/niezhiqi/Desktop/figure/accuracy_plot18_adam.png')
plt.show()
```

```python
#resnet18+adam
import json
import matplotlib.pyplot as plt

# 1. 读取 JSON 文件
def load_json_data(json_path):
    with open(json_path, 'r') as f:
        data = json.load(f)
    return data

# 2. 绘制准确率曲线
def plot_accuracy_curves(json_path, save_path=None):
    # 加载数据
    data = load_json_data(json_path)
    
    # 提取训练和验证准确率
    train_acc = data['train_acc_history']
    val_acc = data['val_acc_history']
    epochs = range(1, len(train_acc) + 1)
    
    # 绘制图像
    plt.figure(figsize=(10, 5))
    plt.plot(epochs, val_acc, 'r-', label='Validation Accuracy')
    
    # 设置标题和标签
    plt.title('Training and Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True)
    
    # 保存图像（可选）
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
    
    plt.show()

# 3. 调用函数绘制图像
json_path = 'path/to/your/accuracy_data.json'  # 替换为你的 JSON 文件路径
plot_accuracy_curves(json_path, save_path='accuracy_plot.png')
```
